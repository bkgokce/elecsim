```python
import pandas as pd
import glob, os
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import table
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
os.chdir("/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/run/batchrunners/scenarios/data/fourth_run_8_iterations_lower_taxes_no_max_investment_2018/")
all_data = []
for file_name in glob.glob('*.csv'):

    scenario_results = pd.read_csv(file_name,dtype=np.float64, index_col=None, header=0)
    scenario_results = scenario_results.rename(index=str, columns = {"Recip_gas":"Recip. Gas"})
    scenario = file_name.split("_")[0]+file_name.split("_")[1]+file_name.split("_")[2]
    scenario_results['Unnamed: 0'] += 2013
    scenario_results = scenario_results.iloc[6:]
    scenario_results['scenario'] = [scenario]*len(scenario_results)
    scenario_results['total'] = scenario_results.iloc[:,1:8].sum(axis=1)
    scenario_results.iloc[:,1:8] = scenario_results.iloc[:,1:8].div(scenario_results.total/100, axis=0)

    scenario_results['Low Carbon'] = scenario_results['Onshore']+scenario_results['Offshore']+scenario_results['PV']+scenario_results['Nuclear']
    scenario_results['Traditional Generation'] = scenario_results['CCGT']+scenario_results['Coal']+scenario_results['Recip. Gas']
    # print(scenario_results)
    all_data.append(scenario_results)
frame = pd.concat(all_data, axis=0, ignore_index=True)

# frame_long = pd.melt(frame, id_vars=['Unnamed: 0','scenario','Carbon_tax'],value_vars=['CCGT', 'Coal', 'Onshore', 'Offshore', 'PV', 'Nuclear','Recip. Gas'])
# frame_long = frame_long.rename(index=str, columns={'variable':"Variables"})
```

```python
frame['Unnamed: 0'] = pd.to_numeric(frame['Unnamed: 0'])
frame
```

```python
bins = [2019, 2029,2039, 2050]
# frame = frame.rename(index=str, columns={"Unnamed: 0":"Year Interval"})
scenario_group = frame.groupby(['scenario', pd.cut(frame['Unnamed: 0'],bins=bins)])
```

```python
summary_stats = scenario_group['Low Carbon','Traditional Generation'].describe().round(2)
# summary_stats = summary_stats.rename(level=2, columns={"Unnamed: 0",'Year Interval'})
summary_stats[['Low Carbon','Traditional Generation']]
summary_stats = summary_stats.drop(["count",'50%','75%','25%'], axis=1, level=1)
summary_stats=summary_stats.reset_index()
summary_stats['Demand'], summary_stats['Carbon Tax'] = summary_stats['scenario'].str.split("-",1).str
summary_stats = summary_stats.drop(columns="scenario")
summary_stats['Demand'] = summary_stats['Demand'].str.replace("demand099","Demand Decreasing 1% a Year")
summary_stats['Demand'] = summary_stats['Demand'].str.replace("demand101","Demand Increasing 1% a Year")
summary_stats['Carbon Tax'],_ = summary_stats['Carbon Tax'].str.split('-').str
_,summary_stats['Carbon Tax'] = summary_stats['Carbon Tax'].str.split('n').str
summary_stats['Carbon Tax'] = summary_stats['Carbon Tax'].str.replace("178",'170 to 22')
summary_stats['Carbon Tax'] = summary_stats['Carbon Tax'].str.replace("18",'26 to 174')
summary_stats
summary_stats = summary_stats[['Demand','Carbon Tax','Unnamed: 0','Low Carbon','Traditional Generation']]
summary_stats = summary_stats.set_index(['Demand', "Carbon Tax",'Unnamed: 0'])
summary_stats
```
