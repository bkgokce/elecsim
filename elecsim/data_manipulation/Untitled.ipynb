```python
import pandas as pd
import matplotlib.pyplot as plot
pd.set_option('display.max_rows', 5000)
```

```python
pp_dat = pd.read_csv('/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/data/processed/power_plants/uk_power_plants/power_plant_db_with_simplified_type.csv')
```

```python
pp_dat.Start_date.iloc[418]=2002
pp_dat.Start_date = pd.to_numeric(pp_dat.Start_date)
pp_dat
```

```python
print(pp_dat.Simplified_Type.unique())
print(len(pp_dat.Simplified_Type.unique()))
```

```python
generator_types = pp_dat.Simplified_Type.unique()

fig=plot.figure(figsize=(16, 16))

for i, var_name in enumerate(generator_types):
#     ax = fig. add_subplot(5,3,i+1)
    df = pp_dat[pp_dat.Simplified_Type==var_name]
    fig.add_subplot(5,3, i+1)
    plot.hist(df.Start_date)
    plot.title(var_name)
    #     ax.set_title(var_name)
    
fig.tight_layout()
plot.show()
```
