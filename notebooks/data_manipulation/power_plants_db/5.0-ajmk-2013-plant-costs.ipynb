```python
def simplified_type_converter(row):
    row_plant = row['Plant_Type']
    if "CCGT" in row_plant and "CCS" in row_plant:
        return "CCGT_CCS"
    elif "CCGT" in row_plant:
        return "CCGT"
    elif "Coal" in row_plant and "CCS" in row_plant:
        return "Coal"
    elif "OCGT" in row_plant and "CCS" in row_plant:
        return "OCGT_CCS"
    elif "OCGT" in row_plant:
        return "OCGT"
    elif "Nuclear" in row_plant:
        return "Nuclear"
    elif "ACT" in row_plant:
        return "ACT"
    elif "biomass" in row_plant:
        return "Biomass"
    elif "Biomass" in row_plant:
        return "Biomass"
    elif "Onshore" in row_plant:
        return "Onshore"
    elif "Offshore" in row_plant:
        return "Offshore"
    elif "Hydro" in row_plant:
        return "Hydro"
    elif "Energy from Waste" in row_plant:
        return "EfW"
    elif "AD" in row_plant:
        return "AD"
    elif "Solar" in row_plant:
        return "PV"
    elif "solar" in row_plant:
        return "PV"
    else:
        return np.NaN

dat['Type'] = dat.apply(lambda x: simplified_type_converter(x), axis=1)
dat[['Plant_Type','Type']]
```
